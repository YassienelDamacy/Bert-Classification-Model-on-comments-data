### Removing URLs

```python
# function that removes urls from data
def remove_urls(dataframe, column_name):
    # put the style of url 
    url_pattern = r'https?://\S+|www\.\S+'
    
    # Function inside that removes URLs and replace it with space 
    def remove_url(text):
        return re.sub(url_pattern, ' ', text)
    
    # Apply the remove_url function to the column and moves on all rows
    dataframe[column_name] = dataframe[column_name].apply(lambda x: remove_url(x))
    
    return dataframe
```

### Return words to it's basic form

```python
# manual function that return Short Expression to it's basic form and also common punctuation in data and replace it with space
def remove(text):
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"can't", "cannot ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'scuse", " excuse ", text)
    text = re.sub('\W', ' ', text)
    text = re.sub('\s+', ' ', text)
    text = re.sub('\n', '  ', text)
    text = text.strip(' ')
    return text
```

```python
# apply function on mentioned column 
df['comment_text'] = df['comment_text'].map(lambda x : remove(x))
```

### Remove punctuation 

```python
# function that removes punctuation and replace it with space
def remove_punctuation(text):
    
    # replace punctuation with space
    translator = str.maketrans('', '', string.punctuation)
    
    # clean data from punctuation
    cleaned_text = text.translate(translator)
    
    return cleaned_text
```

## Remove stop words

```python
# check all stop words that will be removed in english language
stop_words = stopwords.words('english')
stop_words
```

```python
# Function to remove stop words
def remove_stop_words_from_column(text):  
    # tokenizes the  text  into individual words using the  word_tokenize  function. 
    word_tokens = word_tokenize(text)
    
    # collect all words not included in stop_word list by iterating over each word in word_tokens and put it in filtered_text
    filtered_text = [word for word in word_tokens if word not in stop_words]
    
    # Return the filtered text
    return ' '.join(filtered_text)
```

#### count stop words before the function

```python
# Collect all stop words and save it to variable (stop_words)
stop_words = set(stopwords.words('english'))

# Function to check for number of stop words
def count_rows_stop_words(text):
    tokens = nltk.word_tokenize(text)
    return any(token in stop_words for token in tokens)

# Count the number of rows with stop words
rows_with_stop_words = df['comment_text'].apply(count_rows_stop_words).sum()
rows_with_stop_words
```

### Using Word2Vec libarary

```python
# Assuming you have a list of stemmed sentences
stemmed_sentences = df['stemmed_new_column']

# split each sentence to words
sentence_words = [sentence.split() for sentence in stemmed_sentences]

# Train the model
# best range of vector size is between (100-300) which is dimension
# window is distance between current and predicted word 
# workers is umber of CPU cores to use for training
model = Word2Vec(sentence_words, vector_size=300, window=5, min_count=1, workers=4)

# Get the word embeddings for all words
word_embeddings = model.wv
```

```python
# Collect word embeddings in a list
embedding_list = []

# List to store the corresponding words
word_list = []

# iterate over each word in data
for word in word_embeddings.key_to_index:
    # put each score of word in a variable
    embedding = word_embeddings[word]

    # then append it in an array
    embedding_list.append(embedding)

    # apeend the word in an array to be the index
    word_list.append(word)
```

```python
# Create a dataframe from the embedding list with words as index
word_embedding_df = pd.DataFrame(embedding_list, index=word_list)

word_embedding_df
```

```python
# make new dataframe containing only needed columns for bert model
new_df = df.loc[:, ['comment_text','offensive']]

new_df
```

### Remove some rows from data 

#### why remove some rows from data ? because if we used all the data the model will run in many hours and will exceed the deadline 

```python
# will remove the 40 % of data 
percentage_to_remove = 0.4
rows_to_remove = int(len(new_df) * percentage_to_remove)

# Randomly select rows to remove
random.seed(42)  
rows_to_remove_indices = random.sample(range(len(new_df)), rows_to_remove)

# new dataframe by reseting the index
df_filtered = new_df.drop(rows_to_remove_indices).reset_index()
```

```python
# drop column index
df_filtered.drop('index',axis=1,inplace=True)
```

```python
df_filtered
```

### keras 

```python
# Split the data into input text and labels of the column offensive and get value as numpy array
text = df_filtered['comment_text'].values
labels = df_filtered['offensive'].values

# Convert labels to binary matrix
labels = to_categorical(labels)
```

```python
# Split the data train and test 
train_text, test_text, train_labels, test_labels = train_test_split(text, labels, test_size=0.2, random_state=42)
```

```python
# download pretrained model for BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = TFBertModel.from_pretrained('bert-base-uncased')
```

```python
# Maximum sequence length for BERT
max_length = 128  

# tokenize the text train and test data
train_encodings = tokenizer.batch_encode_plus(train_text.tolist(), max_length=max_length, padding=True, truncation=True)
test_encodings = tokenizer.batch_encode_plus(test_text.tolist(), max_length=max_length, padding=True, truncation=True)

# np.array(encodings['input_ids']) NumPy array contains the tokenized input sequences converted into integer
# np.array(train_encodings['attention_mask']) NumPy array contains the attention mask 
train_inputs = [np.array(train_encodings['input_ids']), np.array(train_encodings['attention_mask'])]
test_inputs = [np.array(test_encodings['input_ids']), np.array(test_encodings['attention_mask'])]
```

```python
# input_ids and attention_mask are the input layers in the neural networks and input length is 128
input_ids = Input(shape=(max_length,), dtype=tf.int32, name="input_ids")
attention_mask = Input(shape=(max_length,), dtype=tf.int32, name="attention_mask")

# put the input layers inn the pretrained model
outputs = bert_model(input_ids, attention_mask=attention_mask)

# extracts the pooled output from the BERT model output
pooled_output = outputs[1]

# output layer is 2 nodes with softmax as an activation function
output = Dense(2, activation='softmax')(pooled_output)  

# define model input and output
model = Model(inputs=[input_ids, attention_mask], outputs=output)
```

```python
# compile using ADAM optimizer and loss function is cross entropy
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# fit 1 epoch and batch size equal 16
model.fit(train_inputs, train_labels, epochs=1, batch_size=16)  
```

```python
# predict using test data
predictions = model.predict(test_inputs)

# get the higger probabilty to assign the sentence to it from predicted data
predicted_labels = np.argmax(predictions, axis=1)

# get the higger probabilty to assign the sentence to it from actual data
true_labels = np.argmax(test_labels, axis=1)

# print classification report
print(classification_report(true_labels, predicted_labels))
```

#### explaination of the output
- Precision : measure of how many of the predicted positive instances were actually positive
- Recall : measure of how many of the actual positive instances were correctly predicted as positive
- F1-Score : harmonic mean of precision and recall
- Support : number of instances in the true dataset that belong to each class.
- Accuracy : the accuracy of model which is 89 % 
- Macro Avg : These values are the unweighted average of precision, recall, and F1-score for all classes 
- Sample Avg : These values are similar to the micro average but consider each sample equally.
